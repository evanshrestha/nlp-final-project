```python
import glob

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
data = []

for file in glob.glob("*_run_*.txt"):
    metric_file = f"{file}_metrics.json"
    # Not safe
    dict_metric = eval(open(metric_file).read())
    embedding, tokenization, is_aug, _, run = file.split('.')[0].split('_')
    
    data.append((embedding, tokenization, int(is_aug == 'aug'), int(run), dict_metric['EM'], dict_metric['F1']))
    
data_df = pd.DataFrame(data, columns=['embedding', 'tokenization', 'is_aug', 'run_num', 'em', 'f1'])
```

```python
data_df.groupby(['embedding', 'tokenization', 'is_aug']).agg({'em': ['mean', 'sem'], 'f1': ['mean', 'sem']})
```
