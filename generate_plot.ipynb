```python
import glob
import subprocess

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
data = []

for file in sorted([x for x in glob.glob("model_output/*_run_*.txt") if '0.' not in x]):
    metric_file = f"{file}_metrics.json"
    # Not safe
    dict_metric = eval(open(metric_file).read())
    embedding, tokenization, is_aug, _, run = file.split('/')[1].split('.')[0].split('_')
    
    data.append((embedding, tokenization, int(is_aug == 'aug'), int(run), dict_metric['EM'], dict_metric['F1']))
    
baseline_df = pd.DataFrame(data, columns=['embedding', 'tokenization', 'is_aug', 'run_num', 'em', 'f1'])
```

```python
# Aug probability is set to 0.1
baseline_df.groupby(['embedding', 'tokenization', 'is_aug']).agg({'em': ['mean', 'sem'], 'f1': ['mean', 'sem']})
```

```python
data = []

for path in sorted(glob.glob("model_output/*_aug_0*.json")):
    # Not safe
    dict_metric = eval(open(path).read())
    embedding, tokenization, _, aug_prob, _, run = path.split('/')[1].split('.txt')[0].split('_')
    
    data.append((embedding, tokenization, float(aug_prob), int(run), dict_metric['EM'], dict_metric['F1']))

ablation_df = pd.DataFrame(data, columns=['embedding', 'tokenization', 'aug_prob', 'run_num', 'em', 'f1']) 
```

```python
ablation_df.groupby(['embedding', 'tokenization', 'aug_prob']).agg({'em': ['mean', 'sem'], 'f1': ['mean', 'sem']})
```
